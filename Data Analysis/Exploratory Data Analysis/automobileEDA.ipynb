Basics of Grouping

```python
df['drive-wheels'].unique()
```

```python
# find which type of drive wheel is most valuable
df_group_one = df[["drive-wheels", "body-style", "price"]]
df_group_one = df_group_one.groupby(["drive-wheels"], as_index=False).mean()
df_group_one.sort_values(by=['price'], ascending=False)
```

```python
df_gptest = df[['drive-wheels', 'body-style', 'price']]
grouped_test1 = df_gptest.groupby(['drive-wheels', 'body-style'], as_index=False).mean()
grouped_test1 = grouped_test1.sort_values(by=['price'], ascending=False)
grouped_test1
```

```python
grouped_pivot = grouped_test1.pivot(index='drive-wheels', columns='body-style')
grouped_pivot
```

```python
grouped_pivot = grouped_pivot.fillna(0)
grouped_pivot
```

```python
# group body-style to find avg price
df_2 = df[['body-style', 'price']]
df_group2 = df_2.groupby(['body-style'], as_index=False).mean()
df_group2.sort_values(['price'], ascending=False, inplace=True)
df_group2
```

Variables: Drive Wheels and Body Style vs. Price

```python
fig, ax = plt.subplots()
im = ax.pcolor(grouped_pivot, cmap='RdBu')

#label names
row_labels = grouped_pivot.columns.levels[1]
col_labels = grouped_pivot.index

#move ticks and labels to the center
ax.set_xticks(np.arange(grouped_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(grouped_pivot.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```

Wheel-Base vs. Price

```python
from scipy import stats
```

```python
pearson_coef, p_value = stats.pearsonr(df["wheel-base"], df["price"])
pearson_coef, p_value
```

```python
# => since p-value < 0.001, the correlation between wheel-base and price is statistically significant, although the linear relationship is not extremely strong
```

Hoursepower vs. Price

```python
pearson_coef, p_value = stats.pearsonr(df["horsepower"], df["price"])
print(
    "The Pearson Correlation Coefficient is",
    pearson_coef,
    " with a P-value of P =",
    p_value,
)
```
